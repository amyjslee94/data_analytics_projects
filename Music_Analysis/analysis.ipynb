```python
# Query to fetch the first 10 rows of the release table
query = """
SELECT i.isrc, l.name, rd.length, t.*
FROM track t
LEFT JOIN medium m 
        ON t.medium = m.id
LEFT JOIN medium_format mf 
        ON m.format = mf.id
LEFT JOIN isrc i 
        ON t.recording = i.recording
LEFT JOIN release r 
        ON m.release = r.id
LEFT JOIN recording rd
        ON t.recording = rd.id
LEFT JOIN language l
        ON r.language = l.id
WHERE i.isrc IS NOT NULL
    AND r.language IS NOT NULL
    AND r.packaging = 89
LIMIT 10
;
"""

# Read query into a pandas DataFrame
try:
    df = pd.read_sql_query(query, conn)
    print("Query executed successfully!")
    display(df)
except Exception as e:
    print(f"Error: {e}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Initialize Chrome WebDriver with WebDriverManager
options = webdriver.ChromeOptions()
driver = webdriver.Chrome()
# Open the target URL
url = "https://charts.youtube.com/charts/TopSongs/kr/weekly/20241212"
driver.get(url)

try:
    print("Waiting for the page to fully load...")
    # Wait for the page to load completely by checking for a specific element
    WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.TAG_NAME, "paper-icon-button"))
    )
    print("Page loaded successfully!")

    # Locate all 'paper-icon-button' elements
    buttons = driver.find_elements(By.TAG_NAME, "paper-icon-button")
    # attr = buttons.execute_script('var items = {}; for (index = 0; index < arguments[0].attributes.length; ++index) { items[arguments[0].attributes[index].name] = arguments[0].attributes[index].value }; return items;', buttons)    
    
    # Loop through the buttons and print their attributes
    for i, button in enumerate(buttons):
        if button.get_attribute('title') == 'download':
            is_visible = driver.execute_script("return arguments[0].offsetParent !== null;", button)
            print(f"Is button visible? {is_visible}")
            if is_visible:
                # Trigger click using JavaScript
                button.click()
            else:
                print("Button is not visible or interactable.")
        else:
            print(button.get_attribute('title'))

except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # Close the WebDriver
    driver.quit()
```
